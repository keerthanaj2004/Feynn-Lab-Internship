# Does Population Density impact the Sales of the EV?

**IMPORTING NEW DATASET- POPULATION PER STATE**

```python
popu=pd.read_excel("State wise distribution of population.xlsx")
```

```python
popu.head()
```

```python

columns_to_keep = ['Name', 'Population Density per sq km']
popu = popu[columns_to_keep]
```

```python
popu.head()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(data=popu, y=popu['Name'].sort_values(ascending=True), x='Population Density per sq km', palette='viridis')
plt.ylabel('State')
plt.xlabel('Population density')
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.show()
```

```python
sales["State Name"].unique()
```

```python
popu["Name"].unique()
```

```python
popu=popu[popu["Name"]!='India']
```

```python
condition = popu['Name'] == 'NCT of Delhi'
popu.loc[condition, 'Name'] = 'Delhi'
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(data=popu, y=popu['Name'].sort_values(ascending=True), x='Population Density per sq km', palette='viridis')
plt.ylabel('State')
plt.xlabel('Population density')
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.show()
```

```python
popu.rename(columns={'Name': 'State Name'}, inplace=True)
```

```python
merged_df = pd.merge(popu, sales, on='State Name')
merged_df['Population Density per sq km'].max()
```

```python
states_above_2000_density = merged_df[merged_df['Population Density per sq km'] == 11297.0]['State Name']
for state_name in states_above_2000_density:
    print(state_name)
```

```python
merged_df=merged_df[merged_df['Grand Total']>0]
```

```python
sns.lineplot(data=merged_df,x="Population Density per sq km",y="Grand Total",estimator='mean', errorbar=None)
```

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python

df_encoded = pd.get_dummies(merged_df, columns=['State Name'])
X = df_encoded['Population Density per sq km'].values.reshape(-1, 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

n_clusters = 3

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)

merged_df['Cluster'] = kmeans.labels_

segment_analysis = merged_df.groupby('Cluster').agg(lambda x: x.mean() if x.dtype.kind in 'biufc' else x)

print(segment_analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=segment_analysis.reset_index(), x='Cluster', y='Population Density per sq km', palette='viridis')
plt.xlabel('Cluster')
plt.ylabel('Mean Feature1')
plt.title('Cluster Profiles - Mean Feature1')
plt.show()
```
